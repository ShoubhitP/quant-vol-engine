## Section 1: Itô's Lemma and the Black-Scholes PDE

### Itô's Lemma

For a stochastic process $dS = \mu S\,dt + \sigma S\,dW$ and a function $V(S,t)$, Itô's Lemma states:

$$dV = \frac{\partial V}{\partial t}dt + \frac{\partial V}{\partial S}dS + \frac{1}{2}\frac{\partial^2 V}{\partial S^2}(dS)^2$$

### Step 1: Compute $(dS)^2$

$$dS = \mu S\,dt + \sigma S\,dW$$

$$(dS)^2 = (\mu S\,dt + \sigma S\,dW)^2 = \mu^2S^2(dt)^2 + 2\mu\sigma S^2(dt)(dW) + \sigma^2S^2(dW)^2$$

Applying the stochastic calculus rules:
- $(dt)^2 \to 0$ (higher order, negligible)
- $(dt)(dW) \to 0$ (higher order, negligible)  
- $(dW)^2 \to dt$ (quadratic variation of Brownian motion)

Therefore:
$$(dS)^2 = \sigma^2S^2\,dt$$

### Step 2: Substitute into Itô's Lemma

$$dV = \frac{\partial V}{\partial t}dt + \frac{\partial V}{\partial S}(\mu S\,dt + \sigma S\,dW) + \frac{1}{2}\frac{\partial^2 V}{\partial S^2}(\sigma^2S^2\,dt)$$

### Step 3: Group dt and dW terms

$$\boxed{dV = \left(\frac{\partial V}{\partial t} + \mu S\frac{\partial V}{\partial S} + \frac{1}{2}\sigma^2S^2\frac{\partial^2 V}{\partial S^2}\right)dt + \sigma S\frac{\partial V}{\partial S}\,dW}$$

### Step 4: Eliminate Risk via Portfolio Hedging

Construct a portfolio $\Pi = V - \Delta S$ where $\Delta = \frac{\partial V}{\partial S}$. The dW terms cancel:

$$d\Pi = \left(\frac{\partial V}{\partial t} + \frac{1}{2}\sigma^2S^2\frac{\partial^2 V}{\partial S^2}\right)dt$$

By no-arbitrage, $d\Pi = r\Pi\,dt = r(V - \Delta S)dt$. Setting equal and rearranging gives the **Black-Scholes PDE**:

$$\boxed{\frac{\partial V}{\partial t} + \frac{1}{2}\sigma^2S^2\frac{\partial^2 V}{\partial S^2} + rS\frac{\partial V}{\partial S} - rV = 0}$$

# Black-Scholes Derivation from the Heat Equation
We are deriving the Black-Scholes formula from the heat equation, because the BS PDE comes from the heat equation through a change of variables. 
From the heat equation, you can solve it analytically, and that solution is the closed form of the BS PDE, allowing us to price options.

$$\frac{\partial V}{\partial t} + \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2} + rS\frac{\partial V}{\partial S} - rV = 0$$

- ∂V/∂t — how the option value changes over time (theta decay)
- ½σ²S²∂²V/∂S² — the convexity term, how the option value curves with respect to stock price (related to gamma)
- rS∂V/∂S — drift of the stock price under the risk-neutral measure (related to delta)
- rV — discounting the option value at the risk-free rate

In a perfectly hedged portfolio, the return must equal the risk-free rate. This is the no-arbitrage condition baked into BS.

## Section 2: Change of Variables

We make the following substitutions to simplify the BS Partial Differential Equation:

- $x = \log(S/K)$ — log-moneyness (removes the S terms)
- $\tau = T - t$ — time to expiry (reverses time direction)  
- $u(x, \tau) = e^{r\tau} V$ — discounted option value

These substitutions transform the awkward S² terms into constants, allowing the BS PDE to reduce to the standard heat equation.

# - $x = \ln(S/K)$ → $S = Ke^x$
- $\tau = T - t$ → $\frac{\partial \tau}{\partial t} = -1$
- $V = e^{-r\tau}u$ → $u = e^{r\tau}V$

**Step 1: Compute $\frac{\partial V}{\partial S}$**

$$\frac{\partial V}{\partial S} = \frac{\partial V}{\partial x} \cdot \frac{\partial x}{\partial S} = e^{-r\tau}\frac{\partial u}{\partial x} \cdot \frac{1}{S}$$

So: $rS\frac{\partial V}{\partial S} = re^{-r\tau}\frac{\partial u}{\partial x}$

**Step 2: Compute $S^2\frac{\partial^2 V}{\partial S^2}$**

By product rule:
$$\frac{\partial^2 V}{\partial S^2} = e^{-r\tau}\left[\frac{\partial^2 u}{\partial x^2}\cdot\frac{1}{S^2} - \frac{\partial u}{\partial x}\cdot\frac{1}{S^2}\right]$$

So: $\frac{1}{2}\sigma^2 S^2\frac{\partial^2 V}{\partial S^2} = \frac{1}{2}\sigma^2 e^{-r\tau}\left[\frac{\partial^2 u}{\partial x^2} - \frac{\partial u}{\partial x}\right]$

**Step 3: Compute $\frac{\partial V}{\partial t}$**

$$\frac{\partial V}{\partial t} = re^{-r\tau}u - e^{-r\tau}\frac{\partial u}{\partial \tau}$$

**Step 4: Substitute into BS PDE and divide by $e^{-r\tau}$**

$$ru - \frac{\partial u}{\partial \tau} + \frac{1}{2}\sigma^2\left[\frac{\partial^2 u}{\partial x^2} - \frac{\partial u}{\partial x}\right] + r\frac{\partial u}{\partial x} - ru = 0$$

The $ru$ terms cancel:

$$\frac{\partial u}{\partial \tau} = \frac{1}{2}\sigma^2\frac{\partial^2 u}{\partial x^2} + \left(r - \frac{1}{2}\sigma^2\right)\frac{\partial u}{\partial x}$$

**Step 5: Eliminate drift with $\xi = x + k\tau$, where $k = r - \frac{1}{2}\sigma^2$**

Let $u(x,\tau) = w(\xi, \tau)$. Then:
- $\frac{\partial u}{\partial \tau} = k\frac{\partial w}{\partial \xi} + \frac{\partial w}{\partial \tau}$
- $\frac{\partial u}{\partial x} = \frac{\partial w}{\partial \xi}$
- $\frac{\partial^2 u}{\partial x^2} = \frac{\partial^2 w}{\partial \xi^2}$

Substituting:
$$k\frac{\partial w}{\partial \xi} + \frac{\partial w}{\partial \tau} = \frac{1}{2}\sigma^2\frac{\partial^2 w}{\partial \xi^2} + k\frac{\partial w}{\partial \xi}$$

The $k\frac{\partial w}{\partial \xi}$ terms cancel, leaving:

$$\boxed{\frac{\partial w}{\partial \tau} = \frac{1}{2}\sigma^2\frac{\partial^2 w}{\partial \xi^2}}$$

**This is the heat equation. Black-Scholes is heat diffusion in disguise.**

## Section 2: Girsanov's Theorem and Risk-Neutral Pricing

- Girsanov's theorem allows us to shift into a more risk-neutral measure which makes the mathematics much easier
- The measure P has drift term mu, but by transforming it into measure Q, we can change the drift term into risk-free r
- λ, specifically, is the drift adjustment factor, and allows us to transform from measure P with drift term μ into measure Q with r
- Under the real world measure P, a stock earns μ (which includes a risk premium for taking on risk). Girsanov strips that risk premium away. Under Q, the stock only earns r, the same as putting money in a bank. No extra reward for risk. Since everything earns r under Q, the option payoff grows at r. So to find what it's worth today, you just discount at r.

### Key Equations

Under the real-world measure $\mathbb{P}$:
$$dS = \mu S\,dt + \sigma S\,dW^{\mathbb{P}}$$

Under the risk-neutral measure $\mathbb{Q}$:
$$dS = rS\,dt + \sigma S\,dW^{\mathbb{Q}}$$

The market price of risk:
$$\lambda = \frac{\mu - r}{\sigma}$$

The Radon-Nikodym derivative (change of measure):
$$\frac{d\mathbb{Q}}{d\mathbb{P}} = e^{-\lambda W^{\mathbb{P}}_T - \frac{1}{2}\lambda^2 T}$$

Risk-neutral pricing formula:
$$\text{Option Price} = e^{-rT}\,\mathbb{E}^{\mathbb{Q}}[\text{Payoff}]$$

This is exactly what our Monte Carlo implementation computes:
$$\text{price} = e^{-rT} \cdot \frac{1}{N}\sum_{i=1}^{N} \max(S_T^{(i)} - K, 0)$$

## Section 3: Black-Scholes Assumption Stress Test

Black-Scholes is built on five key assumptions. Below we stress-test each one against real market behavior.

### Assumption 1: Constant Volatility
**BS assumes:** $\sigma$ is fixed and constant across all strikes and expiries.

**Reality:** Our implied volatility surface directly disproves this. After extracting IV from live option chains and fitting SVI parametrization, the surface shows clear volatility smiles; IV is highest for deep OTM puts, lower ATM, and rises again for OTM calls. IV also varies significantly across expiries. A single constant $\sigma$ cannot capture this behavior.

### Assumption 2: Log-Normal Returns
**BS assumes:** Stock returns follow Geometric Brownian Motion, implying normally distributed log-returns.

**Reality:** Real financial markets have fat tails. Extreme events occur far more frequently than a normal distribution predicts. Black Monday (1987) was a 20+ standard deviation event under log-normal assumptions, which should be essentially impossible. Heston addresses this partially by allowing volatility to be stochastic, which generates heavier tails.

### Assumption 3: Constant Risk-Free Rate
**BS assumes:** $r$ is constant over the life of the option.

**Reality:** The Federal Reserve changes interest rates continuously. In 2022, rates rose from 0% to 5% in a single year. Any BS option price computed at $r = 0\%$ at the start of 2022 was materially wrong by year end.

### Assumption 4: No Dividends
**BS assumes:** The underlying asset pays no dividends.

**Reality:** When a stock pays a dividend, the stock price drops by the dividend amount on the ex-dividend date. Standard BS ignores this, causing call prices to be systematically overestimated for dividend-paying stocks like Apple. In practice, traders adjust using a continuous dividend yield $q$, modifying the BS formula to:

$$C = Se^{-qT}N(d_1) - Ke^{-rT}N(d_2)$$

### Assumption 5: Continuous Trading
**BS assumes:** The hedging portfolio can be rebalanced continuously with no transaction costs.

**Reality:** Markets close overnight, on weekends, and during crashes. You cannot rebalance your delta hedge when the market is closed. This gap risk means the perfectly hedged portfolio BS assumes is impossible in practice. Transaction costs further erode the replication argument.

---

**Key takeaway:** Black-Scholes is not wrong; It is a tractable approximation that works well for liquid, short-dated, near-ATM options. Its failure modes are well understood, which is precisely why we built Heston and the implied volatility surface: to model the behavior that BS cannot capture.

## Section 4: Feynman-Kac and the Black-Scholes Formula

### The Feynman-Kac Theorem

Feynman-Kac connects two completely different mathematical objects — a PDE and a probabilistic expectation. For the Black-Scholes PDE:

$$\frac{\partial V}{\partial t} + rS\frac{\partial V}{\partial S} + \frac{1}{2}\sigma^2S^2\frac{\partial^2 V}{\partial S^2} - rV = 0$$

With terminal condition $V(S,T) = \text{Payoff}(S_T)$, Feynman-Kac states the solution is:

$$V(S,t) = e^{-r(T-t)}\mathbb{E}^{\mathbb{Q}}[\text{Payoff}(S_T) | S_t = S]$$

For a call option:

$$C(S,t) = e^{-rT}\mathbb{E}^{\mathbb{Q}}[\max(S_T - K, 0)]$$

This is exactly what our Monte Carlo computes. In other words, **Monte Carlo is numerical Feynman-Kac.**

---

### Deriving the Black-Scholes Formula from Feynman-Kac

Under $\mathbb{Q}$, $S_T$ is log-normally distributed:

$$S_T = Se^{(r-\frac{1}{2}\sigma^2)T + \sigma\sqrt{T}Z}, \quad Z \sim N(0,1)$$

The expectation becomes:

$$C = e^{-rT}\int_{-\infty}^{\infty} \max\left(Se^{(r-\frac{1}{2}\sigma^2)T + \sigma\sqrt{T}z} - K,\ 0\right) \cdot \frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}dz$$

### Step 1: Find the integration limit

The max is nonzero when $S_T > K$. Setting $S_T = K$ and solving for $z^*$:

$$Se^{(r-\frac{1}{2}\sigma^2)T + \sigma\sqrt{T}z^*} = K$$

$$z^* = \frac{\ln(K/S) - (r-\frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}} = -d_2$$

Where:
$$d_2 = \frac{\ln(S/K) + (r-\frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}}$$

### Step 2: Split the integral

$$C = e^{-rT}\int_{-d_2}^{\infty} Se^{(r-\frac{1}{2}\sigma^2)T + \sigma\sqrt{T}z} \cdot \frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}dz - Ke^{-rT}\int_{-d_2}^{\infty} \frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}dz$$

### Step 3: Evaluate the second integral

$$Ke^{-rT}\int_{-d_2}^{\infty} \frac{1}{\sqrt{2\pi}}e^{-\frac{z^2}{2}}dz = Ke^{-rT} \cdot P(Z > -d_2) = Ke^{-rT}N(d_2)$$

### Step 4: Evaluate the first integral by completing the square

Focus on the exponent:
$$\sigma\sqrt{T}z - \frac{z^2}{2} = -\frac{1}{2}(z - \sigma\sqrt{T})^2 + \frac{1}{2}\sigma^2T$$

Substituting back and simplifying ($e^{-\frac{1}{2}\sigma^2T} \cdot e^{\frac{1}{2}\sigma^2T} = 1$):

$$Se^{-rT} \cdot e^{rT}\int_{-d_2}^{\infty} \frac{1}{\sqrt{2\pi}}e^{-\frac{1}{2}(z-\sigma\sqrt{T})^2}dz$$

Substituting $w = z - \sigma\sqrt{T}$, lower limit becomes $-d_2 - \sigma\sqrt{T} = -d_1$:

$$S\int_{-d_1}^{\infty} \frac{1}{\sqrt{2\pi}}e^{-\frac{w^2}{2}}dw = S \cdot N(d_1)$$

Where:
$$d_1 = d_2 + \sigma\sqrt{T} = \frac{\ln(S/K) + (r+\frac{1}{2}\sigma^2)T}{\sigma\sqrt{T}}$$

### Step 5: The Black-Scholes Formula

$$\boxed{C = SN(d_1) - Ke^{-rT}N(d_2)}$$

---

### The Complete Derivation Chain

| Step | Tool | Result |
|------|------|--------|
| 1 | Itô's Lemma | $dV$ in terms of $dS$ and $dt$ |
| 2 | Portfolio hedging + no-arbitrage | Black-Scholes PDE |
| 3 | Change of variables | Heat equation |
| 4 | Girsanov's Theorem | Risk-neutral measure $\mathbb{Q}$ |
| 5 | Feynman-Kac | PDE solution = expectation under $\mathbb{Q}$ |
| 6 | Log-normal integration | Black-Scholes formula $C = SN(d_1) - Ke^{-rT}N(d_2)$ |
